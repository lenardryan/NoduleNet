```python
(input, truth_bboxes, truth_labels, truth_masks, mask, image) = (dataset[i][0], dataset[i][1], dataset[i][2], dataset[i][3], dataset[i][4], dataset[i][5])
```

### Get Predictions

```python
threshold = 0.75

try:
  D, H, W = image.shape
  pid = dataset.filenames[i]

  print('[%d] Predicting %s' % (i, pid), image.shape)
  gt_mask = mask.astype(np.uint8)

  with torch.no_grad():
      input = input.cuda().unsqueeze(0)
      net.forward_pred(input)

  rpns = net.rpn_proposals.cpu().numpy()
  detections = net.detections.cpu().numpy()
  ensembles = net.ensemble_proposals.cpu().numpy()

  if len(detections) and net.use_mask:
      crop_boxes = net.crop_boxes
      # detections_crop_boxes = np.delete(detections[detections[:,1]>threshold], 1, 1).astype(int)
      detections_new = detections[detections[:,1]>threshold,1:8]
      segments = [F.sigmoid(m).cpu().numpy() > 0.5 for m in net.mask_probs]
      # segments = [m.cpu().numpy() for m in net.mask_probs]
      segments_2 = [m.cpu().numpy() for m in net.mask_probs]

      pred_mask = crop_boxes2mask(crop_boxes[detections[:,1]>threshold, 1:], segments, input.shape[2:])
      pred_mask = pred_mask.astype(np.uint8)

      pred_mask_2 = crop_boxes2mask(crop_boxes[detections[:,1]>threshold, 1:], segments_2, input.shape[2:])
      pred_mask_2 = pred_mask_2.astype(np.uint8)
  else:
      pred_mask = np.zeros((input[0].shape))

except Exception as e:
  del input, truth_bboxes, truth_labels, truth_masks, mask, image,
  torch.cuda.empty_cache()
  traceback.print_exc()
              
  print
```

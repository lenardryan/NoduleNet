```python
import os
from os.path import exists
import pandas as pd
import csv

all_file = open("/scratch1/joseph.portugal/NoduleNet/split/3.csv", "r")
all_data = list(csv.reader(all_file, delimiter=","))

train_file = open("/scratch1/joseph.portugal/NoduleNet/split/3_train.csv", "r")
train_data = list(csv.reader(train_file, delimiter=","))

test_file = open("/scratch1/joseph.portugal/NoduleNet/split/3_val.csv", "r")
test_data = list(csv.reader(test_file, delimiter=","))

preprocessed_data_dir = '/scratch1/joseph.portugal/Luna16/dataset/preprocessed_test/3'

for all_filename, train_filename, test_filename in zip(all_data, train_data, test_data):
    
    all_file_exists = exists(os.path.join(preprocessed_data_dir, '%s_bboxes.npy' % all_filename))
    if not all_file_exists:
        all_data.remove(all_filename)
    
    train_file_exists = exists(os.path.join(preprocessed_data_dir, '%s_bboxes.npy' % train_filename))
    if not train_file_exists:
        train_data.remove(train_filename)
    
    test_file_exists = exists(os.path.join(preprocessed_data_dir, '%s_bboxes.npy' % test_filename))
    if not test_file_exists:
        test_data.remove(test_filename)    
    
all_df = pd.DataFrame(all_data)
all_df.to_csv('/scratch1/joseph.portugal/NoduleNet/split/new/3.csv', index=False, header=False)

train_df = pd.DataFrame(train_data)
train_df.to_csv('/scratch1/joseph.portugal/NoduleNet/split/new/3_train.csv', index=False, header=False)

test_df = pd.DataFrame(test_data)
test_df.to_csv('/scratch1/joseph.portugal/NoduleNet/split/new/3_val.csv', index=False, header=False)
```
